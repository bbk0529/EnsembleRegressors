```python
from regressors import *
import pickle
import pandas as pd
```

# Data Setup

```python
#parameter setup
n_stations = 450
n_timesteps= 100 #maybe 144 is the maximum to propoerly manage them 
k = round(n_timesteps/10)
p_noise_stations = 0.05
p_noise_timesteps= 0.05
min_noises = 5
max_noises = 8
```

```python
data = Tempearture_DWD(n_stations, n_timesteps, k, p_noise_stations=p_noise_stations, p_noise_timesteps= p_noise_timesteps, min_noises = min_noises, max_noises = max_noises)
```

```python
ts_data = data.ts_data
neighbor = data._neighbor
```

```python
station = np.random.choice(range(n_stations))
plt.plot(data._ts_rawdata[station])
plt.plot(data._ts_rawdata[neighbor[station]].T, color='grey')
```

# Noises Exploration

```python
noises = data.matrix_noises.flatten()[abs(data.matrix_noises.flatten()) > 0]
plt.hist(noises, bins= (max_noises)*2)
```

```python
# Parameter Setup / Data 
```

```python
#instantiate objects


smoothpredictor = SmoothingAndPredict()
spatialcomparision = SpatialComparision()

executorNew = Executor(data, smoothpredictor)
executorOld = Executor(data, spatialcomparision)
```

```python
resultNew = executorNew.evaluate_validator()
[resultNew['precision'], resultNew['recall'], resultNew['f1'], resultNew['runtime']]
```

```python
resultOld = executorOld.evaluate_validator()
[resultOld['precision'], resultOld['recall'], resultOld['f1'], resultOld['runtime']]
```

# Test with a station randomly selected 

```python
station = np.random.choice(list(data.dic_timesteps.keys()))
idx = (neighbor[station] >= np.inf )
for s in neighbor[station] : 
    if s in list(data.dic_timesteps.keys()) : 
        idx = idx | (neighbor[station] == s)        
        
print(station, np.sum(idx))
```

```python
y = copy.deepcopy(ts_data[station])
X = copy.deepcopy(ts_data[neighbor[station]])
smoothpredictor.correct(X, y,graph=True)
```

```python
y = ts_data[station]
X = ts_data[neighbor[station]]

plt.figure(figsize=(20,3))
plt.plot(X.T, color='grey')
plt.plot(y, color='blue')
reg = LinearRegression().fit(X.T,y)
score = reg.score(X.T, y)
print(score)
plt.plot(reg.predict(X.T), color='red')
```

```python
plt.figure(figsize=(10,2))
plt.bar(range(k), reg.coef_)
plt.figure(figsize=(10,2))
plt.bar(range(k), idx, alpha=0.2, color='red')
```

```python
executorNew.validate(station)
```

```python

try : 
    answer = sorted(data.dic_timesteps[station])
except :
    answer = []

pred_by_new = [int(s.split('_')[1]) for s in executorNew.validate(station)]
pred_by_old = [int(s.split('_')[1]) for s in executorOld.validate(station)]

print("selected station", station)
print("neighbor stations with noise: ", [s for s in data._neighbor[station] if s in data._picked_stations])
print("answer:         ",answer)
print("predicted by new",pred_by_new, levenshteinDistance(answer, pred_by_new ))
print("predicted by old",pred_by_old, levenshteinDistance(answer, pred_by_old ))
plt.hist(data.matrix_noises[station][abs(data.matrix_noises[station]) >0], bins=20 )
```

```python
value_plot = max(abs(ts_data[station]))
plt.figure(figsize=(20,5))
plt.plot(ts_data[neighbor[station]].T, color='grey')
plt.plot(ts_data[station], color='blue', linewidth=2)
# plt.plot(data._ts_rawdata[station], color='orange')
plt.title(station)

plt.figure(figsize=(20,5))
plt.plot(data._ts_rawdata[station] + 0.5, color='grey')
plt.plot(ts_data[station], color='blue')
plt.plot(smoothpredictor.correct(ts_data[neighbor[station]], ts_data[station]) -0.5, color='red')



print(value_plot)
bar1 = np.zeros(len(ts_data[station]))
bar2 = np.zeros(len(ts_data[station]))
bar1[answer] = value_plot
bar2[pred_by_new] = value_plot
plt.bar(np.arange(len(bar1)), bar1, color='red', alpha=0.4)
plt.bar(np.arange(len(bar1)), -bar2, color='green', alpha=0.4)
# plt.legend(['data_w_noise','original','predict'])
```
