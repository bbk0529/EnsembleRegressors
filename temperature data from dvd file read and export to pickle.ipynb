```python
import os
```

```python
import pandas as pd
```

```python
from pathlib import Path
RESULT = []
for i in os.listdir('.') :
    subpath = i
    try : 
        files_in_basepath = os.listdir(subpath)
        for item in files_in_basepath:
            filename = "./" + subpath +"/" + item
            print(filename)
            if filename.endswith('txt') : 
                df=pd.read_csv(filename, sep=';')
                RESULT.append(df)
    except : 
        continue

    
    
```

```python
data = pd.concat(RESULT,axis=0)
```

```python
import pickle
pickle.dump(data, open('data_dvd.p','wb'))
```
